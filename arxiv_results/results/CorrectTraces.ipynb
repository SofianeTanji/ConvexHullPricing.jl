## Utilitaries

```julia
# Imports and utility functions
using JLD2, Plots, StatsBase, ConvexHullPricing
const UT = ConvexHullPricing.Utilitaries
const OPT = ConvexHullPricing.Optimizer
RANGE = 0:1:900
function refine_it(valuevec, timevec)
    new_values = Float64[]
    for t in RANGE
        push!(new_values, valuevec[searchsortedfirst(timevec, t)])
    end
    return new_values
end

function get_relative_error(array, idx)
    precision = Float64[]
    for elt in array
        push!(precision, ((elt)/abs(Opts[idx])))
    end
    return threshold.(precision)
end
function get_relative_error_CA(array, idx)
    precision = Float64[]
    for elt in array
        push!(precision, 100 * (elt/abs(OptsCA[idx])))
    end
    return precision
end
function make_it_monotone(array)
    new_array = Float64[]
    push!(new_array, array[1])
    for elt in array[2:end]
        if elt < minimum(new_array)
            push!(new_array, elt)
        else
            push!(new_array, minimum(new_array))
        end
    end
    return (new_array)
end
```

```julia
function refine_it_less(valuevec, timevec)
  new_values = Float64[]
  for t in RANGE
    x = searchsortedfirst(timevec, t)
    if x > length(valuevec)
      push!(new_values, new_values[end])
    else
      push!(new_values, valuevec[x])
    end
  end
  return new_values
end
```

```julia
function threshold(x)
    return maximum([x, 0])
end
```

```julia
function who_is_best_at_error(err)
  CG = searchsortedfirst(MonPlotCG, err, rev=true)
  BPLM = searchsortedfirst(MonPlotBPLM, err, rev=true)
  BLM = searchsortedfirst(MonPlotBLM, err, rev=true)
  ESTPOL = searchsortedfirst(MonPlotESTPOL, err, rev=true)
  DA = searchsortedfirst(MonPlotDA, err, rev=true)
  DOWG = searchsortedfirst(MonPlotDOWG, err, rev=true)
  RSG = searchsortedfirst(MonPlotRSG, err, rev=true)
  SG = searchsortedfirst(MonPlotnSubg, err, rev=true)
  lSG = searchsortedfirst(MonPlotlastSubG, err, rev=true)
  GM = searchsortedfirst(MonPlotGM, err, rev=true)
  FGM = searchsortedfirst(MonPlotFGM, err, rev=true)
  if minimum([CG, BPLM, BLM, ESTPOL, DA, DOWG, RSG, SG, lSG, GM, FGM]) == CG
    return "Column Generation"
  end
  if minimum([CG, BPLM, BLM, ESTPOL, DA, DOWG, RSG, SG, lSG, GM, FGM]) == BPLM
    return "Bundle Proximal Level Method"
  end
  if minimum([CG, BPLM, BLM, ESTPOL, DA, DOWG, RSG, SG, lSG, GM, FGM]) == BLM
    return "Bundle Level Method"
  end
  if minimum([CG, BPLM, BLM, ESTPOL, DA, DOWG, RSG, SG, lSG, GM, FGM]) == ESTPOL
    return "Subgradient Method with estimated Polyak stepsizes"
  end
  if minimum([CG, BPLM, BLM, ESTPOL, DA, DOWG, RSG, SG, lSG, GM, FGM]) == DA
    return "D-Adaptation"
  end
  if minimum([CG, BPLM, BLM, ESTPOL, DA, DOWG, RSG, SG, lSG, GM, FGM]) == DOWG
    return "DoWG"
  end
  if minimum([CG, BPLM, BLM, ESTPOL, DA, DOWG, RSG, SG, lSG, GM, FGM]) == RSG
    return "Restarted Subgradient Method with vanishing step lengths"
  end
  if minimum([CG, BPLM, BLM, ESTPOL, DA, DOWG, RSG, SG, lSG, GM, FGM]) == SG
    return "Subgradient Method with vanishing step lengths"
  end
  if minimum([CG, BPLM, BLM, ESTPOL, DA, DOWG, RSG, SG, lSG, GM, FGM]) == lSG
    return "Last-iterate Optimal Subgradient Method"
  end
  if minimum([CG, BPLM, BLM, ESTPOL, DA, DOWG, RSG, SG, lSG, GM, FGM]) == GM
    return "Gradient Descent"
  end
  if minimum([CG, BPLM, BLM, ESTPOL, DA, DOWG, RSG, SG, lSG, GM, FGM]) == FGM
    return "Fast Gradient Descent"
  end
end
```

```julia
function rank_them_all(err)
  CG = searchsortedfirst(MonPlotCG, err, rev=true)
  BPLM = searchsortedfirst(MonPlotBPLM, err, rev=true)
  BLM = searchsortedfirst(MonPlotBLM, err, rev=true)
  ESTPOL = searchsortedfirst(MonPlotESTPOL, err, rev=true)
  DA = searchsortedfirst(MonPlotDA, err, rev=true)
  DOWG = searchsortedfirst(MonPlotDOWG, err, rev=true)
  RSG = searchsortedfirst(MonPlotRSG, err, rev=true)
  SG = searchsortedfirst(MonPlotnSubg, err, rev=true)
  lSG = searchsortedfirst(MonPlotlastSubG, err, rev=true)
  GM = searchsortedfirst(MonPlotGM, err, rev=true)
  FGM = searchsortedfirst(MonPlotFGM, err, rev=true)
  bests = Dict(
    "Column Generation" => CG,
    "Bundle Proximal Level Method" => BPLM,
    "Bundle Level Method" => BLM,
    "Subgradient Method with estimated Polyak stepsizes" => ESTPOL,
    "D-Adaptation" => DA,
    "DowG" => DOWG,
    "Restarted Subgradient Method with vanishing step lengths" => RSG,
    "Subgradient Method with vanishing step lengths" => SG,
    "Last-iterate Optimal Subgradient Method" => lSG,
    "Gradient Descent" => GM,
    "Fast Gradient Method" => FGM
  )
  sorted_dict = sort(collect(bests), by=x -> x[2])
  ranked_dict = Dict()
  for (rank, (key, _)) in enumerate(sorted_dict)
    ranked_dict[key] = rank
  end
  return ranked_dict
end
```

```julia
function vectRTA()
  Results = Dict(
    "Column Generation" => Int64[],
    "Bundle Proximal Level Method" => Int64[],
    "Bundle Level Method" => Int64[],
    "Subgradient Method with estimated Polyak stepsizes" => Int64[],
    "D-Adaptation" => Int64[],
    "DowG" => Int64[],
    "Restarted Subgradient Method with vanishing step lengths" => Int64[],
    "Subgradient Method with vanishing step lengths" => Int64[],
    "Last-iterate Optimal Subgradient Method" => Int64[],
    "Gradient Descent" => Int64[],
    "Fast Gradient Method" => Int64[]
  )
  for err in ErrorRange
    classement = rank_them_all(err)
    for (key, rank) in classement
      push!(Results[key], rank)
    end
  end
  return Results
end
```
